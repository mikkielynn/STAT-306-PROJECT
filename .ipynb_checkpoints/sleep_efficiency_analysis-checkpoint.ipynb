### Reading dataset into R ###

```r
# Loading libraries, packages
library(tidyverse)
library(dplyr)
library(cowplot)
library(ggplot2)
```

```r
sleep <- read.table("Sleep_Efficiency.csv", sep = ",", header = TRUE)
head(sleep)
```

### Choosing our explanatory variables ###

```r
# Visualizing the relationship between age and sleep efficiency
options(repr.plot.height = 5, repr.plot.width = 10) 
age_plot <- sleep %>% 
    ggplot(aes(x = Age, y = Sleep.efficiency)) +
    geom_point(aes(colour = Age)) +
    xlab("Age (years)") + 
    ylab("Sleep efficiency (%)")

# Visualizing the relationship between awakenings and sleep efficiency
awakenings_plot <- sleep %>% 
    ggplot(aes(x = Awakenings, y = Sleep.efficiency)) +
    geom_boxplot(aes(group = Awakenings)) +
    xlab("Awakenings (out of 4)") + 
    ylab("Sleep efficiency (%)")

plot_grid(age_plot, awakenings_plot)
```

```r
# Visualizing the relationship between caffeine consumption and sleep efficiency
options(repr.plot.height = 5, repr.plot.width = 7) 
require(gridExtra)

caffeine_plot <- sleep %>% 
    ggplot(aes(x = Caffeine.consumption, y = Sleep.efficiency)) +
    geom_boxplot(aes(group = Caffeine.consumption)) + 
    xlab("Caffeine consumption (out of 5)") + 
    ylab("Sleep efficiency (%)")

# Visualizing the relationship between alcohol consumption and sleep efficiency
alcohol_plot <- sleep %>% 
    ggplot(aes(x = Alcohol.consumption, y = Sleep.efficiency)) +
    geom_boxplot(aes(group = Alcohol.consumption)) +
    xlab("Alcohol consumption (out of 5)") + 
    ylab("Sleep efficiency (%)")

# Visualizing the relationship between exercise frequency and sleep efficiency
exercise_plot <- sleep %>% 
    ggplot(aes(x = Exercise.frequency, y = Sleep.efficiency)) +
    geom_boxplot(aes(group = Exercise.frequency)) +
    xlab("Exercise Frequency (out of 5)") + 
    ylab("Sleep efficiency (%)")
exercise_plot

options(repr.plot.height = 7, repr.plot.width = 7) 
grid.arrange(caffeine_plot, alcohol_plot)
```

From the visualizations above, we will focus on testing the relationship between alcohol consumption, exercise frequency, and awakenings against sleep efficiency. We do not see an evident relationship between caffeine consumption and age against sleep efficiency, therefore we chose to exclude these two variables.

### Choosing dummy variables ###

Since dummy variables are used to encode levels of categorical variables, we should assign Gender as a dummy variable to differentiate between "Male" and "Female". By adding dummy variables, our model should look like the following:

```r
# Creating dummy variables
sleep$gender_m <- ifelse(sleep$Gender == "Male", 1, 0)
head(sleep)
```

```r
# Creating our model with dummy variables
dummy_model <- lm(Sleep.efficiency ~ Light.sleep.percentage + Deep.sleep.percentage + Age + gender_m, data = sleep)
summary(dummy_model)

# Comparing to the model without dummy variables
no_dummy_model <- lm(Sleep.efficiency ~ Light.sleep.percentage + Deep.sleep.percentage + Age, data = sleep)
summary(no_dummy_model)
```

From the results we conclude that the dummy variable is not a statistically significant variable because its p-value is greater than our significance level (INSERT ALPHA HERE). Therefore we can drop Gender as a predictor for our model since it is not important to predicting sleep efficiency.

### Exploring interactions between variables ###

```r
light_squared <- sleep$Light.sleep.percentage^2
deep_squared <- sleep$Deep.sleep.percentage^2
age_squared <- sleep$Age^2

# Interaction term
model_int1 <- lm(Sleep.efficiency ~ 
                 Light.sleep.percentage + 
                 Deep.sleep.percentage + 
                 Age +
                 Light.sleep.percentage*Deep.sleep.percentage +
                 Age*Light.sleep.percentage +
                 Age*Deep.sleep.percentage, 
                 data = sleep)
summary(model_int1)

# Quadratic interactions
model_int2 <- lm(Sleep.efficiency ~ 
                 Light.sleep.percentage + 
                 Deep.sleep.percentage + 
                 light_squared + 
                 deep_squared + 
                 Light.sleep.percentage*Deep.sleep.percentage, 
                 data = sleep)
summary(model_int2)
```
